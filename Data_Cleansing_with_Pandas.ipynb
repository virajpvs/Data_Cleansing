```python
import pandas as pd
```

```python
# Uploading file
df = pd.read_excel(r'Customer Call List.xlsx')
df.head()
```

```python
# Read 1st five rows of a dataframe(Table)
df.head()
df
```

```python
# Drop Duplicates
df = df.drop_duplicates()
df
```

```python
# Deleting a Colunm
df = df.drop(columns = "Not_Useful_Column")
df.head()
```

```python
# Now clean "Last_Name" column
# default strip() only cleans white spaces
# we need to specify what values should take out

 # df['Last_Name'].str.strip()
 # df['Last_Name'] = df['Last_Name'].str.lstrip('/') # only given values on left side will delete
 # df['Last_Name'] = df['Last_Name'].str.lstrip('...')
 # df['Last_Name'] = df['Last_Name'].str.rstrip('_') # only given values on right side will delete
df['Last_Name'] = df['Last_Name'].str.strip('..._/123')
df
```

```python
# now want to Standardize "Phone_Number" coulumn
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-z0-9]','') #using Reguler expression
df.head()
```

```python
# need to convert this float column in to string
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
# now want to formatted "Phone_Number" coulumn
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10] )

df['Phone_Number'] = df['Phone_Number'].str.replace('nan--','')
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--','')
df
```

```python
# need to split 'Address' coloum in to 3 coloumns
# need to split on ',' mark
df[['Street_Address','State','Zip_Code']] = df['Address'].str.split(',',2,expand = True)
df
```

```python
# now need to format 'Paying Customer' column
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes','Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No','N')

df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes','Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No','N')
df
```

```python
#df = df.replace('N/a','')
#df = df.replace('NaN','')
df = df.fillna('')
df
```

```python
# drop Y incluse cell in Do_Not_Contact column
for x in df.index:
  if df.loc[x, "Do_Not_Contact"] == 'Y':
    df.drop(x, inplace = True)

df
```

```python
# drop Null values of Phone_Number colomn
for x in df.index:
  if df.loc[x, "Phone_Number"] == '':
    df.drop(x, inplace = True)

df
# anothoer way to drop Null values
#df = df.dropna(subset = "Phone_Number ", inplace = True)
```

```python
df = df.reset_index(drop = True) # Reset the Index
```
